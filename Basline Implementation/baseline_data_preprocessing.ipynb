<h3>High Spatial Resolution Satellite Imagery for Solar Panel Detection using Image Segmentation</h3>


[Zenodo Record](https://zenodo.org/records/5171712)

The dataset provides multi-resolution photovoltaic (PV) samples from satellite and aerial imagery for PV panel segmentation. It includes samples from three sources: PV08 (Gaofen-2 and Beijing-2 satellite imagery), PV03 (aerial photography), and PV01 (UAV orthophotos). The dataset contains rooftop PV samples categorized by roof type and ground PV samples divided by land use type. It is designed for training segmentation models for detecting PV panels. More details are available in the data document. For more details, the data document is mentioned on the webpage. 

```python
import matplotlib.pyplot as plt
from PIL import Image

img_path_1 = '/kaggle/input/satellite-imagery-solar-panels/PV01/PV01/PV01_Rooftop_Brick/PV01_324942_1203839.bmp'
label_path_1 = '/kaggle/input/satellite-imagery-solar-panels/PV01/PV01/PV01_Rooftop_Brick/PV01_324942_1203839_label.bmp'

img_path_2 = '/kaggle/input/satellite-imagery-solar-panels/PV01/PV01/PV01_Rooftop_Brick/PV01_324942_1203842.bmp'
label_path_2 = '/kaggle/input/satellite-imagery-solar-panels/PV01/PV01/PV01_Rooftop_Brick/PV01_324942_1203842_label.bmp'

img_1 = Image.open(img_path_1)
label_1 = Image.open(label_path_1)
img_2 = Image.open(img_path_2)
label_2 = Image.open(label_path_2)

fig, axes = plt.subplots(2, 2, figsize=(10, 10))

axes[0, 0].imshow(img_1)
axes[0, 0].axis('off')
axes[0, 0].set_title("Image 1")

axes[0, 1].imshow(label_1)
axes[0, 1].axis('off')
axes[0, 1].set_title("Label 1")

# Second image and its label
axes[1, 0].imshow(img_2)
axes[1, 0].axis('off')
axes[1, 0].set_title("Image 2")

axes[1, 1].imshow(label_2)
axes[1, 1].axis('off')
axes[1, 1].set_title("Label 2")

plt.tight_layout()
plt.show()
```

The images above show the brick rooftop images for 0.1m per pixel spatial resolution alongside their masks.

```python
import os
from sklearn.model_selection import train_test_split

directory_path = '/kaggle/input/satellite-imagery-solar-panels/PV01/PV01/PV01_Rooftop_FlatConcrete'

all_files = [f for f in os.listdir(directory_path) if os.path.isfile(os.path.join(directory_path, f))]

image_files = [f for f in all_files if 'label' not in f]

train_images, test_images = train_test_split(image_files, test_size=0.2, random_state=42)

print(f"Number of training images: {len(train_images)}")
print(f"Number of testing images: {len(test_images)}")
```

This is a sample test-train split on one small subdirectory, namely the 0.1m spatial resolution concrete rooftop images.
