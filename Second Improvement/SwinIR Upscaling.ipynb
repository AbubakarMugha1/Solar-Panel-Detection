```python
import os
import shutil
from PIL import Image
```

```python
# Define paths
source_train_images = "/kaggle/input/satellite-imagery-lahore-dha-phase-6/train/images"
source_test_images = "/kaggle/input/satellite-imagery-lahore-dha-phase-6/test/images"
dest_dir = "/kaggle/working/SwinIR-main/testsets/samples"

# Create destination directory if it doesn't exist
os.makedirs(dest_dir, exist_ok=True)

def resize_and_copy_images(source_dir, dest_dir, size=(512, 512)):
    """Resize images to specified size and copy to destination directory"""
    for filename in os.listdir(source_dir):
        if filename.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.gif')):
            try:
                # Open and resize image
                img_path = os.path.join(source_dir, filename)
                img = Image.open(img_path)
                img = img.resize(size, Image.LANCZOS)  # High-quality downsampling
                
                # Save to destination
                dest_path = os.path.join(dest_dir, filename)
                img.save(dest_path)
                print(f"Processed: {filename}")
            except Exception as e:
                print(f"Error processing {filename}: {str(e)}")

# Process train images
print("Processing train images...")
resize_and_copy_images(source_train_images, dest_dir)

# Process test images
print("\nProcessing test images...")
resize_and_copy_images(source_test_images, dest_dir)

print("\nAll images have been resized to 512x512 and copied to:", dest_dir)
```
